One of your TAs recorded the colours of the cars parked at parking area last Friday. Following are the results:

Colour | Number of Cars
--- | ---
White | 12 cars
Black | 18 cars
Red | 8 cars
Blue  | 5 cars

**Plot the car colour distribution in parking area last Friday**

We need to learn two new concepts to do that:

* list → A data type to hold group of information
* matplotlib → Python 2D plotting library

```python
# Import required library for plotting
import matplotlib.pyplot as plt

# Data to be plotted
colors = ["white", "black", "red", "blue"]
num = [12, 18, 8, 5]

# Bar chart
plt.bar(colors, num)
plt.xlabel("Colors")
plt.ylabel("Number of Cars")
plt.title("Car Color Distribution")
plt.show()
```
